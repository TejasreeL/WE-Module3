```python
import random

class Card:
    def __init__(self, rank, suit):
        self.rank = rank
        self.suit = suit

    def __str__(self):
        return f"{self.rank}{self.suit}"

class Player:
    def __init__(self, name, hand):
        self.name = name
        self.hand = hand.copy()  # Copy to avoid modifying original hand
        self.score = 0  # Add attribute to store player's score

    def get_bid(self, revealed_diamond):
        # Basic strategy based on diamond value
        diamond_value = revealed_diamond.rank
        if diamond_value <= 6:  # Low value diamond
            return self.play_low_medium_card()
        elif diamond_value <= 10:  # Mid value diamond
            return self.play_medium_card()
        else:  # High value diamond
            return self.play_high_card()

    def play_low_medium_card(self):
        # Choose a random card from the lower half of the remaining hand
        midpoint = len(self.hand) // 2
        return random.choice(self.hand[:midpoint])

    def play_medium_card(self):
        # Choose a random card from the middle half of the remaining hand
        hand_size = len(self.hand)
        quarterpoint = hand_size // 4
        return random.choice(self.hand[quarterpoint:3*quarterpoint])

    def play_high_card(self):
        # Choose a random card from the higher half of the remaining hand
        return random.choice(self.hand[len(self.hand) // 2:])

    def remove_card(self, card):
        if card in self.hand:
            self.hand.remove(card)

def simulate_game(player1_name, player2_name, deck):
    # Shuffle deck and separate diamonds
    random.shuffle(deck)
    diamonds = [card for card in deck if card.suit == "♦"]
    player1_hand = [card for card in deck if card.suit != "♦"]
    player2_hand = [card for card in deck if card.suit != "♦"]

    # Initialize players
    player1 = Player(player1_name, player1_hand)
    player2 = Player(player2_name, player2_hand)

    # Play each round
    for i in range(13):
        print(f"\nRound {i+1}:")
        if diamonds:  # Ensure there are diamonds left to reveal
            diamond = diamonds.pop(0)  # Remove and get the next diamond card
            print(f"Revealed diamond: {diamond}")
            # Get bids
            player1_bid = player1.get_bid(diamond)
            player2_bid = player2.get_bid(diamond)
            # Determine winner
            if player1_bid.rank > player2_bid.rank:
                winner = player1
            elif player2_bid.rank > player1_bid.rank:
                winner = player2
            else:
                winner = None  # Tie
            # Award points and remove played cards
            if winner:
                winner.score += diamond.rank
                winner.remove_card(player1_bid)
                winner.remove_card(player2_bid)
            else:
                print("Tie! Points split.")
        else:
            print("No more diamonds to reveal!")

    # Print results
    print("\nGame Over! Scores:")
    print(f"{player1.name}: {player1.score}")
    print(f"{player2.name}: {player2.score}")
    if player1.score > player2.score:
        print(f"{player1.name} wins!")
    elif player2.score > player1.score:
        print(f"{player2.name} wins!")
    else:
        print("It's a tie!")

# Example usage
deck = [Card(rank, suit) for rank in range(2, 15) for suit in ["♠", "♥", "♣", "♦"]]
simulate_game("You", "Agent G", deck)
```
